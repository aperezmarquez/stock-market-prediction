# 1.- DATASET
A alto nivel, trabajaremos con un conjunto de datos tabulares que describe el mercado de valores a lo largo del tiempo.

Cada fila representa una acción en un momento específico, donde "id" es la identificación de la acción y "era" es la fecha.

Las características describen los atributos de la acción (por ejemplo, el ratio Price/Earning) conocidos en una fecha dada, y "target" es una medida de rendimientos a 20 días.

El conjunto de datos está ofuscado, lo que significa que los identificadores de acciones (stocks) subyacentes, los nombres de las características y las definiciones de objetivos son anónimos. Esto hace que podamos trabajar con estos datos de forma gratuita y que puedan modelarse sin ningún conocimiento (¡o sesgo!) del dominio financiero.

**TRAIN TEST**

Primero, agrupamos el dataset por eras y realizamos un split, dividiéndolas en eras_train y eras_test. Con estas separaciones, dividimos el dataset principal en train y test, manteniendo agrupados los valores de cada era.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

def eras_split(df):
    eras = df[['era']]
    eras = eras.drop_duplicates(subset='era', keep='first')

    # Dividimos las eras en el tamaño de cada conjunto
    eras_train, eras_test = train_test_split(eras, test_size=0.2, random_state=42)

    # Creamos los dataframes para train y test
    df_train = pd.merge(train, eras_train, left_on='era', right_on='era', how='inner')
    df_test = pd.merge(train, eras_test, left_on='era', right_on='era', how='inner')

    return df_train, df_test
```

Con las eras ya divididas, creamos el dataset train y test.

```python
df_train, df_test = eras_split(train)

# Separamos las features, variables de input para el modelo
X_train = df_train.drop(['era', 'target'], axis=1).copy()
X_test = df_test.drop(['era', 'target'], axis=1).copy()
```

**SCORING MODELO**

Usaremos el score CORR, para ello definiremos la funcion practica_corr.

```python
from scipy import stats
import numpy as np

def practica_corr(preds, target):
    # clasifica (manteniendo vínculos) y luego gaussianiza las predicciones para estandarizar las distribuciones de predicción
    ranked_preds = (preds.rank(method="average").values - 0.5) / preds.count()
    gauss_ranked_preds = stats.norm.ppf(ranked_preds)
    # centra objetivos alrededor de 0
    centered_target = target - target.mean()
    # Levanta ambas predicciones y apunta a la potencia de 1,5 para acentuar las colas.
    preds_p15 = np.sign(gauss_ranked_preds) * np.abs(gauss_ranked_preds) ** 1.5
    target_p15 = np.sign(centered_target) * np.abs(centered_target) ** 1.5
    # Finalmente devuelve el Pearson correlation
    return np.corrcoef(preds_p15, target_p15)[0, 1]
```

**MEJORES PARÁMETROS**

Para ajustar los parámetros y obtener los mejores resultados, se podría usar un Grid Search, aunque sería muy lento en este caso. En su lugar, podemos optar por Randomized Search o la optimización bayesiana. Esta última es más precisa que Randomized Search, ya que aprovecha los resultados previos para ajustar los parámetros de manera más eficiente.


**OVERFITTING**

El overfitting ocurre cuando un modelo depende excesivamente de ciertas características, como en algunas neuronas. Para prevenirlo, podemos limitar la profundidad de los árboles de decisión (max_depth), ajustar un máximo de nodos por hoja (max_leaf_nodes) y probar distintos números de árboles (n_estimators). Por ello, utilizaremos la optimización bayesiana, que no solo encuentra los mejores parámetros, sino que también ayuda a reducir el riesgo de overfitting.

```python
from tqdm import tqdm

# Personalizamos el RandomForest para añadir una barra que muestre el progreso del entrenamiento
class ProgressRandomForest(RandomForestRegressor):
    def fit(self, X, y):
        with tqdm(total=self.n_estimators, desc="Fit Random Forest", unit="tree") as pbar:
            for i in range(self.n_estimators):
                super().fit(X,y)
                pbar.update(1)
        return self
```

Decidimos no usar el progress bar, ya que afecta al tiempo de ejecución del entrenamiento. Una solución sería implementar joblib para que el progress bar se ejecute en paralelo al entrenamiento, sin impactar la velocidad. Sin embargo, por ahora no resulta necesario. En su lugar, calcularé el tiempo aproximado de entrenamiento entrenando un único árbol y midiendo su tiempo de ejecución.

```python
import optuna
import time
import warnings

warnings.filterwarnings('ignore', category=DeprecationWarning)

def objetivo(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    max_depth = trial.suggest_int('max_depth', 2, 20)
    max_leaf_nodes = trial.suggest_int('max_leaf_nodes', 15, 50)
    max_features = trial.suggest_int('max_features', 2, 6)

    # Activamos el verbose a 1 para ver el progreso de entreno del modelo
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        max_leaf_nodes=max_leaf_nodes,
        max_features=max_features,
    )

    # Calculamos el tiempo aproximado de entrenamiento
    aprox_time = RandomForestRegressor(
        n_estimators=1,
        max_depth=max_depth,
        max_leaf_nodes=max_leaf_nodes,
        max_features=max_features,
    )

    start = time.time()
    aprox_time.fit(X_train, df_train['target'])
    end = time.time()

    # Mostramos por consola la aproximación en minutos calculada
    print(f"Tiempo de ejecución por árbol = {(end - start):.2f} secs\n" +
            f"Tiempo aproximado de entrenamiento = {(((end - start) * n_estimators)/60):.2f} mins\n" + 
            f"Parámetros = [n_estimators: {n_estimators}, max_depth: {max_depth}, max_leaf_nodes: {max_leaf_nodes}, max_features: {max_features}]")

    # Entrenamos el modelo real
    model.fit(X_train,df_train['target'])
    df_test['prediction'] = model.predict(X_test)

    per_era_corr = df_test.groupby("era").apply(lambda x: practica_corr(x["prediction"], x["target"]))

    return per_era_corr.mean()

opt = optuna.create_study(direction="maximize")
opt.optimize(objetivo, n_trials=20)
```

```python
print("Best Parameters: ", opt.best_params)
print("Best CORR Score: ", opt.best_value)
```

El mejor resultado obtenido con el Random Forest es con los siguientes parámetros: n_estimators = 834, max_depth = 17, max_leaf_nodes = 48, max_features = 2.

```python
best_params = opt.best_params
best_model = RandomForestRegressor(**best_params)
best_model.fit(X_train, df_train['target'])
```

# 4.- Mejora e investigación

Aplicaré feature neutralization para intentar mejorar el modelo. Si los resultados son satisfactorios, no realizaré más cambios. De lo contrario, exploraré otros modelos o técnicas para optimizar el desempeño.

**FEATURE NEUTRALIZATION**

Esta técnica se utiliza para ajustar o neutralizar características que puedan causar comportamientos no deseados, logrando un modelo más robusto frente a cambios en los datos y evitando sesgos.

Para aplicarla, primero identificamos las características a neutralizar. Una vez localizadas, se ajustan y, posteriormente, validamos el modelo nuevamente para evaluar cómo afecta al rendimiento.

**LOCALIZACIÓN**

```python
# Feature Importance from model
feature_importance = model.feature_importances_

importance_df = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

importance_df.head(5)
```

```python
# Feature Time Stability
stability = {}
for feature in feature_cols:
    correlations = []
    for era, data in train.groupby('era'):
        corr = data[feature].corr(data['target'])  # Calcula correlación por era
        correlations.append(corr)
    # Guarda la estabilidad de cada feature
    stability[feature] = np.std(correlations)

stability_df = pd.DataFrame({
    'Feature': list(stability.keys()),
    'Stability': list(stability.values())
}).sort_values(by='Stability')

stability_df.head(5)
```

```python
# Selección de fatures
combined_df = importance_df.merge(stability_df, on='Feature')

threshold_importance = combined_df['Importance'].quantile(0.7)  # Top 30% importance
threshold_stability = combined_df['Stability'].quantile(0.7)    # Bottom 30% stability

selected_features = combined_df[
    (combined_df['Importance'] > threshold_importance) &
    (combined_df['Stability'] > threshold_stability)
]['Feature'].values

selected_features
```

**NEUTRALIZACIÓN DE CARACTERÍSTICAS**

Para implementar esta técnica, primero predeciremos algunos valores con un modelo lineal, y con esos datos neutralizaremos las variables. Sin embargo, mejoraremos este enfoque reemplazando el modelo lineal por *Ridge Regression*, que está diseñado para prevenir el sobreajuste de las características.

*Ridge Regression* se basa en un modelo lineal estándar, pero incorpora un término de penalización que depende de los coeficientes del modelo. Esto lo convierte en una herramienta útil para neutralizar el efecto de ciertos factores de mercado, mejorando la estabilidad del modelo frente a variaciones en los datos.

```python
#from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

X_neutralize = train[selected_features].copy()
y_neutralize = train['target']

#lr = LinearRegression()
ridge = Ridge(alpha=1.0)

# Se neutraliza cada una de las features
for feature in selected_features:
    # Entrenamos el modelo lineal
    #lr.fit(X_neutralize[[feature]], y_neutralize)
    ridge.fit(X_neutralize[[feature]], y_neutralize)
    
    # Conseguimos las predicciones
    #predicted_values = lr.predict(X_neutralize[[feature]])
    predicted_values = ridge.predict(X_neutralize[[feature]])
    
    # Le quitamos el valor predecido a la feature a neutralizar
    train[feature] = X_neutralize[feature] - predicted_values

train
```

Para validar la neutralización, entrenamos un nuevo modelo de Random Forest utilizando los mismos parámetros que obtuvimos al aplicar la optimización bayesiana. # 5.- Submission
Para esto, utilizaré la librería pickle, que permite guardar los modelos entrenados y cargarlos nuevamente más tarde. Esto facilita la persistencia del modelo, evitando tener que reentrenarlo cada vez que se necesite utilizarlo.

Los métodos save_model() y load_model() están definidos en el apartado 2, Modeling, después de haber realizado la optimización bayesiana con la librería Optuna. Estos métodos permiten guardar el modelo optimizado y cargarlo posteriormente, facilitando su reutilización sin necesidad de reentrenar el modelo cada vez.
