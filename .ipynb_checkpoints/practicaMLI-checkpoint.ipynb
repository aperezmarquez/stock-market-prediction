## 1.- DATASET

A alto nivel, trabajaremos con un conjunto de datos tabulares que describe el mercado de valores a lo largo del tiempo.

Cada fila representa una acci칩n en un momento espec칤fico, donde "id" es la identificaci칩n de la acci칩n y "era" es la fecha. 

Las caracter칤sticas describen los atributos de la acci칩n (por ejemplo, el ratio Price/Earning) conocidos en una fecha dada, y "target" es una medida de rendimientos a 20 d칤as.

El conjunto de datos est치 ofuscado, lo que significa que los identificadores de acciones (stocks) subyacentes, los nombres de las caracter칤sticas y las definiciones de objetivos son an칩nimos. Esto hace que podamos trabajar con estos datos de forma gratuita y que puedan modelarse sin ning칰n conocimiento (춰o sesgo!) del dominio financiero.

Descargamos los datos hist칩ricos de entrenamiento y echamos un vistazo de cerca.

```python
# Inicializa NumerAPI - el cliente oficial de Python para acceder al API de nuestro conjunto de datos.
from numerapi import NumerAPI
napi = NumerAPI()

# Imprime todos los ficheros del conjunto de datos disponibles para descargar los m치s recientes.
[f for f in napi.list_datasets() if f.startswith("v5.0")]
```

```python
import pandas as pd
import json

# Descarga los datos de entrenamiento y los metadatos de las caracter칤sticas
# Esto tomar치 unos pocos minutos 游꼿
napi.download_dataset("v5.0/train.parquet");
napi.download_dataset("v5.0/features.json");

# Carga solo el conjunto de features "medium" para reducir el uso de memoria y acelerar el entrenamiento del modelo 
# O bien, utiliza el conjunto de features "all" para usar todas las caracter칤sticas
# Para pruebas r치pidas puedes usar "small" (despu칠s usa "medimum")
feature_metadata = json.load(open("v5.0/features.json"))
feature_cols = feature_metadata["feature_sets"]["small"]
train = pd.read_parquet("v5.0/train.parquet", columns=["era", "target"] + feature_cols)

# Reduce la muestra, tomando cada cuatro eras, para reducir el uso de memoria y acelerar el entrenamiento del modelo (toma las eras 0001, 0005, 0009, ...)
# Comenta la l칤nea a continuaci칩n para utilizar todos los datos.
train = train[train["era"].isin(train["era"].unique()[::4])]
# Cada fila representa una acci칩n en un momento espec칤fico {id: stockId, era: fecha, target: medida de retorno, features: atributos de un stock}
train
```

## 2.- MODELING

Entrenamiento del modelo:

Eres libre de utilizar cualquier algoritmo visto en clase. Pero, en este ejemplo, usaremos LGBMRegressor, basado en 치rboles de decisi칩n para predecir variables continuas. Es una opci칩n popular para las predicciones de este tipo. En cuanto a los mejores hiperpar치metros, estos son solo una base, haz un grid.

```python
# https://lightgbm.readthedocs.io/en/latest/pythonapi/lightgbm.LGBMRegressor.html 
import lightgbm as lgb #En mi mac, tuve que instalar libomp: "brew install libomp"

# https://lightgbm.readthedocs.io/en/latest/Parameters-Tuning.html
model = lgb.LGBMRegressor(
  n_estimators=2000,
  learning_rate=0.01,
  max_depth=5,
  num_leaves=2**5-1,
  colsample_bytree=0.1
)

# This will take a few minutes 游꼿
model.fit(
  train[feature_cols],
  train["target"]
);
```

Ahora hagamos algunas predicciones con los datos de validaci칩n para evaluar el rendimiento de nuestro modelo.

```python
# Descarga validation data 
# Esto tomar치 unos pocos minutos 游꼿
napi.download_dataset("v5.0/validation.parquet");

# Carga el validation data, filtrando por data_type == "validation"
validation = pd.read_parquet("v5.0/validation.parquet", columns=["era", "data_type", "target"] + feature_cols) 
validation = validation[validation["data_type"] == "validation"]
del validation["data_type"]

# Reduce la muestra a cada cuarta era para reducir el uso de memoria y acelerar la evaluaci칩n
# Comenta la siguiente l칤nea para usar todos los datos (mayor uso de memoria, inferencia m치s lenta, evaluaci칩n m치s precisa)
validation = validation[validation["era"].isin(validation["era"].unique()[::4])]

# Las eras est치n separadas por una semana, pero los "targets" miran 4 semanas en el futuro, por lo que debemos "embargar" las 4 eras posteriores a nuestra 칰ltima "era" para evitar fuga de datos.
last_train_era = int(train["era"].unique()[-1])
eras_to_embargo = [str(era).zfill(4) for era in [last_train_era + i for i in range(4)]]
validation = validation[~validation["era"].isin(eras_to_embargo)]

# Genera predicciones respecto a las caracter칤sticas de validaci칩n fuera de la muestra.
# Esto tomar치 unos pocos minutos 游꼿
validation["prediction"] = model.predict(validation[feature_cols])
validation[["era", "prediction", "target"]]
```

## 3. Evaluaci칩n del rendimiento:

La m칠trica de puntuaci칩n principal que usaremos se llama CORR (practica_corr), que es una variante espec칤fica de para este caso de uso del coeficiente de correlaci칩n de Pearson.

Esta m칠trica est치 dise침ada para "alinear los incentivos" entre el modelo y el desempe침o de un fondo de inversi칩n. Es decir, un modelo con una buena puntuaci칩n CORR deber칤a ayudar a un inversor a obtener buenos rendimientos.

```python
from scipy import stats
import numpy as np

# La principal m칠trica de scoring:
def practica_corr(preds, target):
    #맊lasifica (manteniendo v칤nculos) y luego gaussianiza las predicciones para estandarizar las distribuciones de predicci칩n
    ranked_preds = (preds.rank(method="average").values - 0.5) / preds.count()
    gauss_ranked_preds = stats.norm.ppf(ranked_preds)
    # centra objetivos alrededor de 0
    centered_target = target - target.mean()
    # Levanta ambas predicciones y apunta a la potencia de 1,5 para acentuar las colas.
    preds_p15 = np.sign(gauss_ranked_preds) * np.abs(gauss_ranked_preds) ** 1.5
    target_p15 = np.sign(centered_target) * np.abs(centered_target) ** 1.5
    # Finalmente devuelve el Pearson correlation
    return np.corrcoef(preds_p15, target_p15)[0, 1]
```

Como se mencion칩 anteriormente, es importante calificar cada era hist칩rica de forma independiente. Entonces, al evaluar el desempe침o de nuestro modelo, deber칤amos observar la correcci칩n "por era".

Una cosa que puedes ver aqu칤 es c칩mo de bajas son las puntuaciones (en el rango de +/- 5% de correlaci칩n). Esto es muy normal en el 치mbito de las finanzas cuantitativas y es parte de la raz칩n por la que se indica que es la "pr치ctica de aprendizaje autom치tico m치s dif칤cil" del mundo.


```python
# Calcula la correlaci칩n por era entre nuestras predicciones y los valores "target".
per_era_corr = validation.groupby("era").apply(lambda x: practica_corr(x["prediction"], x["target"]))

# Gr치fica de correlaci칩n por "era" 
per_era_corr.plot(kind="bar", title="Validation Correlation", figsize=(10, 6), xticks=[], snap=False);
```

En lugar de observar la puntuaci칩n "corr" de cada "era", resulta 칰til observar la puntuaci칩n "corr" acumulativa.

Esto se denomina "backtesting" en finanzas cuantitativas, donde las personas simulan el desempe침o hist칩rico de sus estrategias de inversi칩n, puede considerar este gr치fico como una prueba retrospectiva del desempe침o de tu modelo durante el per칤odo de validaci칩n hist칩rica.

```python
# Gr치fica de la correlaci칩n acumulada por "era"
per_era_corr.cumsum().plot(kind="line", title="Cumulative Validation Correlation", figsize=(10, 6));
```

M칠tricas de rendimiento:
    
Para evaluar el rendimiento de nuestro modelo, tambi칠n es 칰til calcular algunas m칠tricas resumidas durante todo el per칤odo de validaci칩n.

La media de correlaciones es la medida principal del rendimiento de su modelo.

Sharpe es una medida de la coherencia de su modelo, un concepto tomado de las finanzas, donde generalmente se refiere a los rendimientos ajustados al riesgo de una estrategia de inversi칩n. Aqu칤 calcularemos Sharpe como la correlaci칩n promedio dividida por la desviaci칩n est치ndar de las correlaciones.

La reducci칩n m치xima (max_drawdown) es una medida del riesgo de tu modelo, otro concepto tomado de las finanzas, donde generalmente se refiere a la p칠rdida financiera m치xima sufrida por una estrategia de inversi칩n. Aqu칤 calcularemos la reducci칩n m치xima como la ca칤da m치xima desde el pico hasta el m칤nimo en la correlaci칩n de validaci칩n acumulativa.

```python
# Calcula m칠tricas de rendimiento
corr_mean = per_era_corr.mean()
corr_std = per_era_corr.std(ddof=0)
corr_sharpe = corr_mean / corr_std
max_drawdown = (per_era_corr.cumsum().expanding(min_periods=1).max() - per_era_corr.cumsum()).max()

pd.DataFrame({
    "mean": corr_mean,
    "std": corr_std,
    "sharpe": corr_sharpe,
    "max_drawdown": max_drawdown
}, index=["Value"]).T
```

Estas m칠tricas de rendimiento anteriores no son sorprendentes, pero son lo suficientemente buenas como para comenzar!
